```python
train_df = raw_train_df.copy()
test_df = raw_test_df.copy()
```

<h5 id="heading"><span style="background-color:#cefffb; color:#1b4946; display:fill;border-radius:5px; font-family:cursive"> üì¢ Helper Function  </span></h5>

```python
# Function 
def combine_columns(column_df):
    """
    Fill the missing value with empty caractere
    and combine title and author and text columns into single column.
    
    Parametres:
    * column_df (pd.DataFrame): Contain only the tree columns: 'title', 'author', and 'text'.
    
    return: (pd.Serie) contain the combined cleaned columns
    """
    
    # fill the missing value with empty caractere
    column_df = column_df.fillna('')
    
    # The text column has larg volume of words
    # if you want to also preprocess it you can let it
    # else you can get rid of it
    column_df['content'] = column_df['author']+' '+column_df['title']+column_df['text']

    return column_df
```

```python
X = combine_columns(train_df)
Y = combine_columns(test_df)
```

```python
len(X), len(Y)
```

```python
X_train = X['content']
X_label =  X['label']
```

```python
Y_test = Y['content']
```

<h3 id="heading"><span style="background-color:#fcf3b9; color:#3d3710 ; display:fill;border-radius:5px; font-family:Georgia">‚öóÔ∏è Prepare the data & Eliminate the stopwords </span></h3>



<span style="font-family:Georgia;
             color:#ff9d91">To clean the missing values, we can do two approaches</span>
             
* <font style="font-family:Georgia; color:#174f20;">

Doing Stemming Process to clean our Data

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re
# Create an instance of PorterStemmer
stemmer = PorterStemmer()

def custom_stemming_function(text):
    # Remove non-alphabetical characters and replace them with spaces
    cleaned_text = re.sub('[^a-zA-Z]', ' ', text)
    
    # Convert the text to lowercase
    lowercase_text = cleaned_text.lower()
    
    # Tokenize the text into words
    words = lowercase_text.split()
    
    # Apply stemming and remove stopwords
    stemmed_words = [stemmer.stem(word) for word in words if word not in stopwords.words('english')]
    
    # Join the cleaned and stemmed words back into a single string
    preprocessed_text = ' '.join(stemmed_words)
    
    return preprocessed_text

X_train=X_train.apply(lambda x: custom_stemming_function(x))
```

```python
X_train.shape
```

```python
Y_test=Y_test.apply(lambda x: custom_stemming_function(x))
```

```python
Y_test.shape
```

```python
train_df = pd.concat([X_train, X['label']], axis=1)
```

```python
test_df = pd.concat([Y_test, raw_subm_df['label']], axis=1)
```
